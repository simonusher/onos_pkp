```python
import networkx as nx
import collections
import matplotlib.pyplot as plt
```

```python
try:
    import pygraphviz
    from networkx.drawing.nx_agraph import graphviz_layout
except ImportError:
    try:
        import pydot
        from networkx.drawing.nx_pydot import graphviz_layout
    except ImportError:
        raise ImportError("This example needs Graphviz and either "
                          "PyGraphviz or pydot")
```

```python
path = 'ic.edges'
f = open(path, 'r')
lines = f.readlines()
f.close()
lines = map(lambda x : x.replace(',',' ').replace('\n',''), lines)
lines = list(lines)[1:]
G = nx.parse_edgelist(lines, nodetype = int, data=(('weight',float),), create_using=nx.DiGraph)
nx.info(G) 
```

```python
pos = graphviz_layout(G, prog='sfdp')
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_size=20, alpha=0.2, node_color="blue", with_labels=False)
plt.axis('equal')
plt.show()
```

```python
pos = graphviz_layout(G, prog='twopi')
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_size=20, alpha=0.5, node_color="blue", with_labels=False)
plt.axis('equal')
plt.show()
```

```python
pos = graphviz_layout(G, prog='neato')
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_size=20, alpha=0.5, node_color="blue", with_labels=False)
plt.axis('equal')
plt.show()
```

```python
pos = graphviz_layout(G, prog='dot')
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_size=20, alpha=0.1, node_color="blue", with_labels=False)
plt.axis('equal')
plt.show()
```

```python
pos = graphviz_layout(G, prog='circo')
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_size=20, alpha=0.1, node_color="blue", with_labels=False)
plt.axis('equal')
plt.show()
```
